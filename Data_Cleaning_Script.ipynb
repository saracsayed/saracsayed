```
# Read in utility block scores csv
block_scores = pd.read_csv('/content/drive/MyDrive/MP/utility_block_scores_2018.csv')
block_scores.head()
```

```
# Read in census block groups csv
census_blocks = pd.read_csv('/content/drive/MyDrive/MP/race_blockgroup_geoid_2018.csv')
census_blocks.head()
```

```
## Read in labor hours data
labor_hours = pd.read_csv('/content/drive/MyDrive/MP/utility_service_scores_2018.csv')
labor_hours.head()
```

```
# Read in rural/urban data
urban_rural = pd.read_csv('/content/drive/MyDrive/MP/UrbanRural_2010.csv')
urban_rural.head()
```

```
# Join block score groups csv with labor hours csv
HBI_race = pd.merge(block_scores, labor_hours, left_on='pwsid', right_on='pwsid')
HBI_race.head()
```

```
## Read in ethnicity data
ethnicity_blocks = pd.read_csv('/content/drive/MyDrive/MP/NotHispanic.csv')
ethnicity_blocks.head(30)
```

## B. Data Modifications to Block Group Demographic Data

```
## Calculate total number of hispanics 
ethnicity_blocks['TotalH'] = ethnicity_blocks['Total H']- ethnicity_blocks['Total NH']
ethnicity_blocks.head(3)
```

```
# Join ethnicity with race csv
race_ethnicity_merged= pd.merge(census_blocks, ethnicity_blocks, left_on='GEOID', right_on='GEOID')
race_ethnicity_merged.head()
```

```
#Join fixed blockgroups with rural/urban
HBI_race_rural = pd.merge(HBI_race, urban_rural, left_on='GEOID', right_on='GEOID')
HBI_race_rural.head()
```

```
# Delete redundant columns in race_ethnicity_merged csv
del HBI_race_rural['HBI_y']
del HBI_race_rural['PPI_y']
del HBI_race_rural['EPA_y']
del HBI_race_rural['burden_y']
del HBI_race_rural['utility_name_y']
HBI_race_rural.head()
```

```
# Rename redundant names
HBI_labor = HBI_race_rural.rename(columns={'utility_name_x':'utility_name', 'HBI_x':'HBI','PPI_x':'PPI', 'burden_x':'burden', 'EPA_x':'EPA','utility_name_x':'utility_name'})
HBI_labor.head()
```

```
# Join block score groups csv with census block groups csv
block_scores_merged= pd.merge(HBI_labor, race_ethnicity_merged, left_on='GEOID', right_on='GEOID')
block_scores_merged.head()
```

```
#Create new columns to show percentages for each ethnic group and rural/urban population
block_scores_merged['Perc_White'] = (block_scores_merged['NH White']/ block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Black'] = (block_scores_merged['NH Black']/ block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Native'] = (block_scores_merged['NH Native']/ block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Asian'] = (block_scores_merged['NH Asian']/block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Pacific_Islander'] = (block_scores_merged['NH Pacific Islander']/block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Other'] = (block_scores_merged['NH Other']/block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Multi'] = (block_scores_merged['NH Two or more']/block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Hispanic'] = (block_scores_merged['TotalH']/block_scores_merged['Total Pop'])*100
block_scores_merged['Perc_Urban'] = (block_scores_merged['Urban']/block_scores_merged['Total'])*100
block_scores_merged['Perc_Rural'] = (block_scores_merged['Rural']/block_scores_merged['Total'])*100
block_scores_merged.head()
```

```
# Remove dashes from the PWSID colummn ans strip the white space
block_scores_merged['pwsid'] = block_scores_merged['pwsid'].replace('-', '', regex=True)
```

```
water_rates = pd.read_csv('/content/drive/MyDrive/MP/utility_rates_table_current.csv')
```

```
# Remove dashes from the PWSID colummn ans strip the white space
water_rates['pwsid'] = water_rates['pwsid'].replace('-', '', regex=True)
```

```
# Export final compiled demographic table for future use
block_scores_merged.to_csv('block_scores_geoid.csv')
water_rates.to_csv('water_rates.csv')
```
